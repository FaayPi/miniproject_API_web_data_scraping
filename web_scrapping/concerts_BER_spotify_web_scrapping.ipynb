```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd
from bs4 import BeautifulSoup

# Download the webdriver
service = Service(ChromeDriverManager().install())

# Start the webDriver
driver = webdriver.Chrome(service=service)

url = 'https://open.spotify.com/concerts/location/2950159-Berlin-DE'

# Load the page
driver.get(url)

# Explicitly wait for an element that ensures the page is loaded
try:
    # Wait until the first concert block is loaded
    concert_block_present = WebDriverWait(driver, 100).until(
        EC.presence_of_element_located((By.CLASS_NAME, "wedRUk4eiBo9n9vqXLhU"))
    )
    print("Page fully loaded!")
except:
    print("Element could not be found, the page may not be fully loaded.")

# Get the HTML source code of the page
html_content = driver.page_source

# Use BeautifulSoup to parse the HTML code
soup = BeautifulSoup(html_content, 'html.parser')

# Find the concert data
concerts = []

for concert_block in soup.find_all("a", class_="wedRUk4eiBo9n9vqXLhU"):  # Searching for 'a' with this specific class
    try:
        # Extract artist name
        artist = concert_block.find("h3", class_="e-9812-text encore-text-body-medium TRp4tgG7_cymGIXDevIP").get_text(strip=True)

        # Extract month and day
        time_tag = concert_block.find("time", class_="HYVIJaYYiCTgSCHie70L")
        month = time_tag.find("div", class_="e-9812-text encore-text-body-small").get_text(strip=True)
        day = time_tag.find("div", class_="e-9812-text encore-text-body-medium-bold").get_text(strip=True)

        # Extract location
        location = concert_block.find("span", class_="e-9812-text encore-text-body-small encore-internal-color-text-subdued l93AfoV_qHnMQSQBB5GK").get_text(strip=True)
        
        concerts.append({
            "artist": artist,
            "month": month,
            "day": day,
            "location": location
        })
    except AttributeError as e:
        # If a tag is not found, print error and continue to the next concert
        print(f"Error extracting a concert: {e}")
        continue

# Close the WebDriver
driver.quit()
```

```python
# Convert the 'concerts' list into a pandas DataFrame
df = pd.DataFrame(concerts)

# Optional: Print the DataFrame to check the results
print("\nDataFrame of concerts in Berlin (extracted from Spotify):")

# Display the DataFrame
df
```
