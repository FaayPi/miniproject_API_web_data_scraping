```python
# API for countries and pupulation
import requests
import pandas as pd

url_countries = "https://restcountries.com/v3.1/all"

# send API request named as response
response = requests.get(url_countries)

# if we can get data (response 200), otherwise error code
if response.status_code == 200:
    data = response.json()
    print(f'API works ({response.status_code})')
else:
    print(f'Error ({response.status_code})')
```

```python
countries_data = response.json()

df_population = pd.json_normalize(countries_data)

# only show colums of interest
columns_of_interest = ['name.common', 'capital', 'population', 'area']

df_population = df_population[columns_of_interest]

df_population
```
